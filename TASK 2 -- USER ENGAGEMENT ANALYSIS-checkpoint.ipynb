# ●	Normalize each engagement metric and run a k-means (k=3) to classify customers into three groups of engagement. 

```python
# Extract the features to be normalized
features = user_engagement[['session_frequency', 'total_duration', 'total_traffic']]
```

```python
# Initialize the scaler
scaler = StandardScaler()
```

```python
# Fit and transform the features
scaled_features = scaler.fit_transform(features)
```

```python
# Initialize the KMeans model with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
# Fit the model to the normalized features
user_engagement['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# To review the resulting DataFrame with cluster labels
user_engagement.head()
```

```python
#TO Show cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```

# Visualization of the K-Means clusters in 2D after reducing the features by PCA

```python
# Apply PCA to reduce the features to 2D
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

# Add PCA results to the DataFrame
user_engagement['pca1'] = pca_features[:, 0]
user_engagement['pca2'] = pca_features[:, 1]

# Define a colormap with three colors and attractive edge colors
colors = ['#FF6347', '#4682B4', '#32CD32']  # Tomato, SteelBlue, LimeGreen
edge_colors = ['#8B0000', '#00008B', '#006400']  # DarkRed, DarkBlue, DarkGreen

# Plot the 2D PCA visualization with custom colors and edge colors
plt.figure(figsize=(10, 6))
for i in range(3):
    cluster_data = user_engagement[user_engagement['cluster'] == i]
    plt.scatter(
        cluster_data['pca1'], 
        cluster_data['pca2'], 
        color=colors[i], 
        edgecolor=edge_colors[i],  # Edge color
        label=f'Cluster {i}', 
        alpha=0.8,  # Adjust transparency
        s=100  # Adjust point size
    )

plt.title('2D PCA Visualization of K-means Clusters')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Clusters')
plt.show()
```

# Using the k-means clustering algorithm, group users in k engagement clusters based on the engagement metrics: 

# 1. Prepare the Data

```python
# Extract the features to be normalized
application_features = apps_df[['Gaming_Total_Data', 'Youtube_Total_Data', 'Netflix_Total_Data',
                                'Google_Total_Data', 'Email_Total_Data', 'Social_Media_Total_Data', 'Other_Total_Data']]
```

```python
# Normalize the data

scaler = StandardScaler()
application_features_scaled = scaler.fit_transform(application_features)
```

```python
# Elbow method to find the optimal k
inertia = []
k_values = range(1, 11)
```

```python
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(application_features_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the elbow method graph
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, 'bo-', markersize=8)
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

# What is the optimized value of k (use the elbow method for this)?  

<font size="3">

***The optimized value of K is 3 from the above elbow method plot***

# 2. Apply K-means Clustering with Optimal k i.e. 3

```python
##Assuming optimal_k =3
optimal_k = 3

# Applying k-means with the optimal k
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(application_features_scaled)

# Add the cluster labels to the original DataFrame
apps_df['Cluster'] = clusters

# Display the first few rows with the cluster labels
apps_df.head()

#TO Show cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```

# Visualization of the Application K-Means clusters in 2D after reducing the features by PCA

```python
# Apply PCA to reduce to 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(application_features_scaled)

# Create a DataFrame with PCA results and cluster labels
pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])
pca_df['Cluster'] = clusters

# Plotting
plt.figure(figsize=(10, 8))
for cluster in pca_df['Cluster'].unique():
    cluster_data = pca_df[pca_df['Cluster'] == cluster]
    plt.scatter(cluster_data['PCA1'], cluster_data['PCA2'], label=f'Cluster {cluster}', s=50, edgecolor='k')

plt.title('2D PCA of K-Means Clusters')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

# Visualization of the Applications K-Means clusters in 3D

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
import pandas as pd

# Define colors and edge colors
colors = ['#FF6347', '#4682B4', '#32CD32']  # Tomato, SteelBlue, LimeGreen
edge_colors = ['#8B0000', '#00008B', '#006400']  # DarkRed, DarkBlue, DarkGreen

# Apply PCA to reduce to 3D
pca = PCA(n_components=3)
pca_result = pca.fit_transform(application_features_scaled)
pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2', 'PCA3'])
pca_df['Cluster'] = clusters

# Plotting
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

for i in range(optimal_k):
    cluster_data = pca_df[pca_df['Cluster'] == i]
    ax.scatter(
        cluster_data['PCA1'], 
        cluster_data['PCA2'], 
        cluster_data['PCA3'], 
        color=colors[i], 
        edgecolor=edge_colors[i],  # Edge color
        label=f'Cluster {i}', 
        alpha=0.8,  # Adjust transparency
        s=50  # Adjust point size
    )

ax.set_title('3D Visualization of K-Means Clusters')
ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')

ax.legend(title='Clusters')
plt.show()
```
