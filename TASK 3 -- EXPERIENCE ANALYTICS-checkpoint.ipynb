```python
df1 = df[['MSISDN/Number','Handset Type','Avg RTT DL (ms)','Avg RTT UL (ms)',
                 'TCP DL Retrans. Vol (Bytes)','TCP UL Retrans. Vol (Bytes)',
                 'Avg Bearer TP DL (kbps)','Avg Bearer TP UL (kbps)']]
```

```python
##Filling the numerical columns with the mean value

df1['Avg RTT DL (ms)'].fillna(value=df1['Avg RTT DL (ms)'].mean(), inplace=True)
df1['Avg RTT UL (ms)'].fillna(value=df1['Avg RTT UL (ms)'].mean(), inplace=True)
df1['Avg Bearer TP DL (kbps)'].fillna(value=df1['Avg Bearer TP DL (kbps)'].mean(), inplace=True)
df1['Avg Bearer TP UL (kbps)'].fillna(value=df1['Avg Bearer TP UL (kbps)'].mean(), inplace=True)
df1['TCP DL Retrans. Vol (Bytes)'].fillna(value=df1['TCP DL Retrans. Vol (Bytes)'].mean(), inplace=True)
df1['TCP UL Retrans. Vol (Bytes)'].fillna(value=df1['TCP UL Retrans. Vol (Bytes)'].mean(), inplace=True)
```

```python
##Filling the categorical columns with the mode value

df1['MSISDN/Number'].fillna(value=df1['MSISDN/Number'].mode()[0], inplace=True)
df1['Handset Type'].fillna(value=df1['Handset Type'].mode()[0], inplace=True)
```

```python
# Plot boxplots for each numerical column
numerical_columns = ['Avg RTT DL (ms)', 'Avg RTT UL (ms)', 'TCP DL Retrans. Vol (Bytes)', 
                     'TCP UL Retrans. Vol (Bytes)', 'Avg Bearer TP DL (kbps)', 'Avg Bearer TP UL (kbps)']

plt.figure(figsize=(15, 10))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)  # Adjust the subplot grid to 2x3 for 6 plots
    sns.boxplot(data=df1, y=column)
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

```python
def replace_outliers_with_mean(df, column, z_thresh=3):
    mean = df[column].mean()
    std = df[column].std()
    
    # Define the z-score threshold for identifying outliers
    z_scores = (df[column] - mean) / std
    
    # Replace outliers with the mean
    df[column] = np.where(np.abs(z_scores) > z_thresh, mean, df[column])
    return df

# Apply outlier replacement to all numerical columns
for column in numerical_columns:
    df1 = replace_outliers_with_mean(df1, column)
```

# Aggregate the information per customer

```python
# Calculating average RTT (combining DL and UL)
df1['Avg RTT (ms)'] = df1[['Avg RTT DL (ms)', 'Avg RTT UL (ms)']].mean(axis=1)
```

```python
# Calculating average TCP retransmission volume (combining DL and UL)
df1['Avg TCP Retrans. Vol (Bytes)'] = df1[['TCP DL Retrans. Vol (Bytes)', 'TCP UL Retrans. Vol (Bytes)']].mean(axis=1)
```

```python
# Calculating average throughput (combining DL and UL)
df1['Avg Throughput (kbps)'] = df1[['Avg Bearer TP DL (kbps)', 'Avg Bearer TP UL (kbps)']].mean(axis=1)
```

```python
# Aggregating per customer
agg_df = df1.groupby('MSISDN/Number').agg({
    'Avg TCP Retrans. Vol (Bytes)': 'mean',
    'Avg RTT (ms)': 'mean',
    'Handset Type': lambda x: x.mode()[0],  # Mode for categorical column
    'Avg Throughput (kbps)': 'mean'
}).reset_index()

# Renaming columns for clarity
agg_df.rename(columns={
    'Avg TCP Retrans. Vol (Bytes)': 'Avg TCP Retransmission (Bytes)',
    'Avg RTT (ms)': 'Avg RTT (ms)',
    'Handset Type': 'Handset Type',
    'Avg Throughput (kbps)': 'Avg Throughput (kbps)'
}, inplace=True)

# Display the aggregated DataFrame
agg_df.head()
```

```python
# Sort the DataFrame by 'Avg TCP Retransmission (Bytes)' in descending order
sorted_by_tcp = agg_df.sort_values('Avg TCP Retransmission (Bytes)', ascending=False)

# Get the top 10 TCP values
top_10 = sorted_by_tcp.head(10)[['MSISDN/Number', 'Avg TCP Retransmission (Bytes)']]

# Get the last 10 TCP values
last_10 = sorted_by_tcp.tail(10)[['MSISDN/Number', 'Avg TCP Retransmission (Bytes)']]

# Get the 10 most frequent TCP values and their counts
most_frequent_values = agg_df['Avg TCP Retransmission (Bytes)'].value_counts().head(10).reset_index()
most_frequent_values.columns = ['Avg TCP Retransmission (Bytes)', 'Frequency']

# Merge with the original DataFrame to get details
most_frequent_tcp = agg_df[agg_df['Avg TCP Retransmission (Bytes)'].isin(most_frequent_values['Avg TCP Retransmission (Bytes)'])]
most_frequent_tcp_summary = most_frequent_tcp.groupby('Avg TCP Retransmission (Bytes)').size().reset_index(name='Occurrences')

# Merge with the frequency counts
most_10 = pd.merge(most_frequent_values, most_frequent_tcp_summary, on='Avg TCP Retransmission (Bytes)')

# Print the results
print("Top 10 TCP Values with MSISDN/Number:\n", top_10)
print("\nLast 10 TCP Values with MSISDN/Number:\n", last_10)
print("\n10 Most Frequent TCP Values with Frequency:\n", most_10)
```

# Task 3.4 - Using the experience metrics above, perform a k-means clustering (where k = 3) to segment users into groups of experiences and provide a brief description of each cluster.

```python
# Select the relevant columns for clustering
clustering_df = agg_df[['Avg TCP Retransmission (Bytes)', 'Avg RTT (ms)', 'Avg Throughput (kbps)']]
```

```python
# Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(clustering_df)

# Perform K-Means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
agg_df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Optionally, print the cluster centers for interpretation
print("Cluster Centers:")
print(scaler.inverse_transform(kmeans.cluster_centers_))

# Display the first few rows of the DataFrame with cluster assignments
agg_df.head()
```

```python
# Plot the clusters
plt.figure(figsize=(14, 7))
sns.scatterplot(data=agg_df, x='Avg Throughput (kbps)', y='Avg TCP Retransmission (Bytes)', hue='Cluster', palette='viridis', s=100, edgecolor='black')
plt.title('K-Means Clustering (k=3) on User Experience Metrics')
plt.xlabel('Average Throughput (kbps)')
plt.ylabel('Average TCP Retransmission (Bytes)')
plt.legend(title='Cluster')
plt.show()
```

```python
# Set up the figure and 3D projection
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the clusters in 3D
sc = ax.scatter(
    agg_df['Avg Throughput (kbps)'], 
    agg_df['Avg TCP Retransmission (Bytes)'], 
    agg_df['Avg RTT (ms)'], 
    c=agg_df['Cluster'], 
    cmap='viridis', 
    s=100, 
    edgecolor='black'
)

# Add labels and title
ax.set_title('3D K-Means Clustering (k=3) on User Experience Metrics')
ax.set_xlabel('Average Throughput (kbps)')
ax.set_ylabel('Average TCP Retransmission (Bytes)')
ax.set_zlabel('Average RTT (ms)')

# Add a color bar for the clusters
plt.colorbar(sc, ax=ax, label='Cluster')

plt.show()
```
